```python
import pydicom as dicom
import matplotlib.pyplot as plt
import os
import cv2
import PIL # optional
import pandas as pd
import csv
from PIL import Image
```

```python
# Load DICOM image
ds = dicom.dcmread('/content/6154')
```

```python
# Get image dimensions
width, height = int(ds.Rows), int(ds.Columns)
```

```python
print('Image width:', width)
print('Image height:', height)
```

```python
pixel_array = ds.pixel_array
```

```python
# Plot the pixel array using matplotlib
plt.imshow(pixel_array, cmap=plt.cm.gray)
plt.show()
```
